```python
import nltk
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.tokenize import sent_tokenize, word_tokenize
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
df = pd.read_csv('./Data files/Data_Car_1000.txt', delimiter = "\t")
df
```

```python
df.info(verbose=True)
```

```python
df.corr()
```

```python
df.describe()
```

```python
unique_counts = df.nunique()
unique_counts
```

```python
non_null_counts = df.count()

columns_to_drop = non_null_counts[non_null_counts < 900].index

df_dropped = df.drop(columns=columns_to_drop)
```

```python
df_dropped.info()
```

```python
df_dropped2= df_dropped.dropna(subset=['Series','BadgeDescription','VIN','EngineConfigurationDescription','Colour','Compliance_Date'])
df_dropped2
```

```python
unique_counts = df_dropped2.nunique()
# Step 2: Identify the columns to remove
columns_to_remove = unique_counts[(unique_counts == 1) | (unique_counts == 0) | (unique_counts == 900)].index.values

df_filtered = df_dropped2.drop(columns=columns_to_remove)
```

```python
df_filtered.info(verbose = True)
```

```python
medians = df_filtered.median()
df_filled = df_filtered.fillna(medians)
df_filled.info()
```

```python
correlation_df = df_filled.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correlation_df, annot=True)
plt.show()
```
